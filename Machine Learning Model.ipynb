```python
import pandas as pd
```

```python
df = pd.read_csv("Melbourne_housing_FULL.csv")
```

```python
df=df.dropna(axis=0)
```

```python
df.head()
```

```python
y = df.Price
```

```python
y
```

```python
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
```

```python
x = df[melbourne_features]
```

```python
x
```

```python
df.head()
```

```python
x.describe()
```

```python
from sklearn.tree import DecisionTreeRegressor
melbourne_model = DecisionTreeRegressor(random_state=1)

# Fit model
melbourne_model.fit(x, y)
# Specifying a number for random_state ensures you get the same results in each run.
#This is considered a good practice. You use any number, and model quality won't depend meaningfully on exactly what value you choose.
```

```python
print("Making predictions for the following 5 houses:")
print(x.head())
print("The predictions are")
print(melbourne_model.predict(x.head()))
```

```python
# Part 2  - tips.csv
tips_data = pd.read_csv('tips.csv')
tips_data.head()
```

```python
# describing the data frame
tips_data.describe()
```

```python
# Removing rows with missing values
tips_data = tips_data.dropna(axis=0)
```

```python
#Specify Prediction Target
y = tips_data.tip
y
```

```python
# changing female and male to zero and one
tips_data['sex'].replace(['Female','Male'],[0,1],inplace = True)
tips_data
```

```python
#Predicting Feautures
feature_names =['total_bill','sex','size','price_per_person']
X = tips_data[feature_names]
X
```

```python
from sklearn.tree import DecisionTreeRegressor
tips_model = DecisionTreeRegressor(random_state=1)
```

```python
# Fitting  the model
tips_model.fit(X,y)
```

```python
print("Making Prediction for the following 5 customers: ")
print(X.head())
print("Predicted tips for the 5 customers: ")
print(tips_model.predict(X.head()))
```
